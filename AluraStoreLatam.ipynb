# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt

# Agrupar por producto y contar
productos_t1 = tienda.groupby("Producto").size().sort_values(ascending=False)
productos_t2 = tienda2.groupby("Producto").size().sort_values(ascending=False)
productos_t3 = tienda3.groupby("Producto").size().sort_values(ascending=False)
productos_t4 = tienda4.groupby("Producto").size().sort_values(ascending=False)
```

```python
def plot_top_bottom(productos, tienda_nombre):
    fig, axs = plt.subplots(1, 2, figsize=(14, 5))

    # Top 5
    productos.head(5).plot(kind='bar', ax=axs[0], color='green')
    axs[0].set_title(f'Top 5 productos más vendidos - {tienda_nombre}')
    axs[0].set_ylabel('Cantidad de ventas')
    axs[0].tick_params(axis='x', rotation=45)

    # Bottom 5
    productos.tail(5).plot(kind='bar', ax=axs[1], color='red')
    axs[1].set_title(f'Top 5 productos menos vendidos - {tienda_nombre}')
    axs[1].set_ylabel('Cantidad de ventas')
    axs[1].tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()

# Mostrar gráficos para cada tienda
plot_top_bottom(productos_t1, "Tienda 1")
plot_top_bottom(productos_t2, "Tienda 2")
plot_top_bottom(productos_t3, "Tienda 3")
plot_top_bottom(productos_t4, "Tienda 4")
```
